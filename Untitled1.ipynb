```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
start= cv2.imread('20210612_051850_Original.jpg')
```

```python
start_rgb=np.copy(start)
```

```python
start_rgb=cv2.cvtColor(start_rgb,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(start_rgb)
```

```python
star_bw=np.copy(start)
```

```python
star_bw=cv2.cvtColor(star_bw,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(star_bw,cmap='gray')
```

```python
cv2.minMaxLoc(star_bw)
```

```python
max_loc_x=cv2.minMaxLoc(star_bw)[3][0]
```

```python
max_loc_y=cv2.minMaxLoc(star_bw)[3][1]
```

```python
s=25
coord1=(max_loc_x+s,max_loc_y+s)
coord2=(max_loc_x-s,max_loc_y-s)
color=(255,0,0)
t=5
brightest_star_box=np.copy(start_rgb)
brightest_star_box=cv2.rectangle(brightest_star_box,coord1,coord2,color,t)
plt.imshow(brightest_star_box)
```
